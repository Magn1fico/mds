```python
import numpy as np
from scipy.optimize import approx_fprime
from sympy import *
```

```python
x,y = symbols('x y')
```

```python
# первая функция

f1 = 2*x**2*y**3 + 1/x + y**2*x + 7 
f1
```

```python
# производная по х

diff(f1,x)
```

```python
# производная по у

diff(f1,y)
```

```python
'градиент в точке  (1,2)'

def f(x):
    return 2*x[0]**2*x[1]**3 + 1/x[0] + x[1]**2*x[0] + 7 
```

```python
x= np.array ([1,2])

approx_fprime(x, f,1e-6)
```

```python
'Уровень 1:'
'Градиентный спуск своими руками:'

def f(x):
    return (x[0]**2) + (x[1]**2)
```

```python
x = np.array([100, 200])
learning_rate = 0.1
```

```python
for i in range(100):
    # Calculate gradient
    grad = approx_fprime(x, f,1e-6)
    
    # Update x with gradient
#     x = np.array([x[0]-learning_rate, x[1]-2*learning_rate])
    x = np.array([x[0]-learning_rate*grad[0], x[1]-learning_rate*grad[1]])
    
    print (grad, x)
```

```python
print("Минимум достигается в: ", np.round(x,2))
print("Значение функции в минимуме: ", round(f(x),2))
```

```python
from scipy.optimize import minimize
```

```python
'Проверка с помощью встроенной функции numpy:'

x = [100, 200]
res = minimize(f, x, method='Nelder-Mead', tol=1e-6)
res
```
