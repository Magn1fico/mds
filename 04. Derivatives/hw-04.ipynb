# –î–æ–º–∞—à–Ω–µ–µ –∑–∞–¥–∞–Ω–∏–µ "–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑. –ü—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è".

## –£—Ä–æ–≤–µ–Ω—å 0:

–ü–æ—Å—á–∏—Ç–∞–π—Ç–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–µ—Ä–≤–æ–π –∏ –≤—Ç–æ—Ä–æ–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$

–í —Ç–æ—á–∫–µ x = 1.

## –£—Ä–æ–≤–µ–Ω—å 1:

–ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫ –ø–µ—Ä–≤–æ–π –∏ –≤—Ç–æ—Ä–æ–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏:

$f(x) = sin^5(2x+1)$

–ù–∞ –æ—Ç—Ä–µ–∑–∫–µ [-5; 5].

```python
'–£—Ä–æ–≤–µ–Ω—å 0:'
'–ü–æ—Å—á–∏—Ç–∞–π—Ç–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–µ—Ä–≤–æ–π –∏ –≤—Ç–æ—Ä–æ–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏:'
'ùëì(ùë•)=ùë•5+4ùë†ùëñùëõ(2ùë•)+ùëêùëúùë†(3ùë•+3)'

import scipy as sc
from scipy import *
import scipy.misc
```

```python
def f(x):
    return x**5 + 4*sin(2*x) + cos(3*x + 3)
```

```python
# –∑–Ω–∞—á–µ–Ω–∏–µ –ø–µ—Ä–≤–æ–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π

sc.misc.derivative(f,1,1e-6)
```

```python
# –∑–Ω–∞—á–µ–Ω–∏–µ –≤—Ç–æ—Ä–æ–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π

sc.misc.derivative(f,1,1e-6,2)
```

```python
'–£—Ä–æ–≤–µ–Ω—å 1:'
'–ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫ –ø–µ—Ä–≤–æ–π –∏ –≤—Ç–æ—Ä–æ–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏:'
'ùëì(ùë•)=ùë†ùëñùëõ5(2ùë•+1)'

from sympy import *

x = symbols('x')
```

```python
def f(x):
    return (sin(2*x+1))**5
```

```python
f1 = diff(f(x),x)
f2 = diff(f1,x)
```

```python
f1 = lambdify(x, f1)
f2 = lambdify(x, f2)
```

```python
import numpy as np
Xf = []
Yf = []
Yf1 = []
Yf2 = []
```

```python
for i in np.linspace(-5,5,100):
    Xf.append(i)
    Yf.append(f(i).evalf())
    Yf1.append(f1(i))
    Yf2.append(f2(i))
```

```python
import matplotlib.pyplot as plt
```

```python
# –≥—Ä–∞—Ñ–∏–∫ –∏—Å—Ö–æ–¥–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ –∏ –ø–µ—Ä–≤–æ–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π

plt.plot(Xf,Yf,label="f(x)")
plt.plot(Xf,Yf1,label="f'(x)")
plt.grid(True)
plt.legend()
```

```python
# –≥—Ä–∞—Ñ–∏–∫ –≤—Ç–æ—Ä–æ–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π

plt.plot(Xf,Yf2)
plt.grid(True)
```
