## Уровень 1:

На отрезке [-5; 5].

```python
'Уровень 0:'
'Посчитайте значение первой и второй производной функции:'
'𝑓(𝑥)=𝑥5+4𝑠𝑖𝑛(2𝑥)+𝑐𝑜𝑠(3𝑥+3)'

import scipy as sc
from scipy import *
import scipy.misc
```

```python
def f(x):
    return x**5 + 4*sin(2*x) + cos(3*x + 3)
```

```python
# значение первой производной

sc.misc.derivative(f,1,1e-6)
```

```python
# значение второй производной

sc.misc.derivative(f,1,1e-6,2)
```

```python
'Уровень 1:'
'Постройте график первой и второй производной функции:'
'𝑓(𝑥)=𝑠𝑖𝑛5(2𝑥+1)'

from sympy import *

x = symbols('x')
```

```python
def f(x):
    return (sin(2*x+1))**5
```

```python
f1 = diff(f(x),x)
f2 = diff(f1,x)
```

```python
f1 = lambdify(x, f1)
f2 = lambdify(x, f2)
```

```python
import numpy as np
Xf = []
Yf = []
Yf1 = []
Yf2 = []
```

```python
for i in np.linspace(-5,5,100):
    Xf.append(i)
    Yf.append(f(i).evalf())
    Yf1.append(f1(i))
    Yf2.append(f2(i))
```

```python
import matplotlib.pyplot as plt
```

```python
# график исходной функции и первой производной

plt.plot(Xf,Yf,label="f(x)")
plt.plot(Xf,Yf1,label="f'(x)")
plt.grid(True)
plt.legend()
```

```python
# график второй производной

plt.plot(Xf,Yf2)
plt.grid(True)
```
